```python
class Route:
    # def __init__(self, next_cell_indices, delta_name, cost):
    def __init__(self, init_pos, cost):
        self.cell_indices = []
        self.delta_names = []
        self.costs = []
        
        self.cell_indices.append(init_pos)
        self.delta_names.append("")
        self.costs.append(cost)
        
    def __str__(self):
        return_string = ""
        
        for cell_pos in self.cell_indices:
            return_string += "[{0:d}, {1:d}] \n".format(cell_pos[0], cell_pos[1])
            
        return return_string
    
    def append(self, cell_pos, delta_name, cost):
        print("appending: ", cell_pos)
        
        self.cell_indices.append(cell_pos)
        self.delta_names.append(delta_name)
        self.costs.append(cost)
        
    def remove(self, cell_pos):
        print("removing: ", cell_pos)
        
        index = self.cell_indices.index(cell_pos)
        
        self.cell_indices.remove(cell_pos)
        self.delta_names.remove(delta_names[index])
        self.costs.remove(costs[index])
    
    def remove_node_up_to(self, cell_pos):
        if (not(cell_pos in self.cell_indices)):
            print("removing up to: ", cell_pos, "  -  ignored.")
            return
        
        cut_index = self.cell_indices.index(cell_pos) + 1
        print("removing up to: ", cell_pos, "  -  with index: ", cut_index)
        
        if (cut_index == len(self.cell_indices)):
            print("not needed")
            
            return
        
        print("before: ", self.cell_indices)
        print("after: ", self.cell_indices[:cut_index])
        
        # excluded_routes_count = len(self.cell_indices) - cut_index
        
        self.cell_indices = self.cell_indices[:cut_index]
        self.delta_names = self.delta_names[:cut_index]
        self.costs = self.costs[:cut_index]
        
        # return excluded_routes_count
        
    def pop(self, log_level = 0):
        
        if (log_level > 0):
            print("popping: ", self.cell_indices[-1])
        
        self.cell_indices.pop()
        self.delta_names.pop()
        self.costs.pop()
        
    def last_pos(self):
        return self.cell_indices[-1]
        
```

```python
def find_neighbor_indices(pos, grid_cells):
    
    # neighbor_poses = []
    
    neighbor_cells_count = 4
    neighbor_indices = []
    delta_names = []
    
    max_size_x = grid_cells.shape[0] - 1
    max_size_y = grid_cells.shape[1] - 1
    
    pos = [min(pos[0], max_size_x), min(pos[1], max_size_y)]
    
    # up
    up_neighbor_indices = [max(pos[0] - 1, 0), pos[1]]
    # down
    down_neighbor_indices = [min(pos[0] + 1, max_size_x), pos[1]]
    # left
    left_neighbor_indices = [pos[0], max(pos[1] - 1, 0)]
    # right
    right_neighbor_indices = [pos[0], min(pos[1] + 1, max_size_y)]
    
    """
    print("up", up_neighbor_indices)
    print("down", down_neighbor_indices)
    print("left", left_neighbor_indices)
    print("right", right_neighbor_indices)
    """
    
    # index = 0
    routs_count = 0
    
    if (pos != up_neighbor_indices):
        
        # trial_grid_bookings[up_neighbor_indices[0], up_neighbor_indices[1]] = 1
        neighbor_indices.append([up_neighbor_indices[0], up_neighbor_indices[1]])
        delta_names.append("^")
        # index += 1
        
        if (grid_bookings[up_neighbor_indices[0], up_neighbor_indices[1]] == 0):
            routs_count += 1
    
    if ((up_neighbor_indices != down_neighbor_indices) and (pos != down_neighbor_indices)):
        
        # trial_grid_bookings[down_neighbor_indices[0], down_neighbor_indices[1]] = 1
        neighbor_indices.append([down_neighbor_indices[0], down_neighbor_indices[1]])
        delta_names.append("v")
        # index += 1
        
        if (grid_bookings[down_neighbor_indices[0], down_neighbor_indices[1]] == 0):
            routs_count += 1
    
    if (pos != left_neighbor_indices):
        
        # trial_grid_bookings[left_neighbor_indices[0], left_neighbor_indices[1]] = 1
        neighbor_indices.append([left_neighbor_indices[0], left_neighbor_indices[1]])
        delta_names.append("<")
        # index += 1
        
        if (grid_bookings[left_neighbor_indices[0], left_neighbor_indices[1]] == 0):
            routs_count += 1
    
    if ((left_neighbor_indices != right_neighbor_indices) and (pos != right_neighbor_indices)):
        
        # trial_grid_bookings[right_neighbor_indices[0], right_neighbor_indices[1]] = 1
        neighbor_indices.append([right_neighbor_indices[0], right_neighbor_indices[1]])
        delta_names.append(">")
        # index += 1
        
        if (grid_bookings[right_neighbor_indices[0], right_neighbor_indices[1]] == 0):
            routs_count += 1
    
    return (neighbor_indices, delta_names, routs_count)
```

```python
def search(grid_cells, grid_bookings, init_pos, goal_pos, cost, max_depth):
    # ----------------------------------------
    # insert code here
    # ----------------------------------------
    
    # route = Route(init_pos, cost)
    
    trial_grid_bookings = np.copy(grid_bookings)
    
    current_pos = init_pos
    all_neighbors_found = False
    branch_search_completed = False
    
    route = Route(init_pos, cost)
    
    debug_counter = 0
    open_routs_count = 0
    last_pos = []
    same_try_count = 0
    
    while (not branch_search_completed):
        # index = 0

        (neighbor_indices, delta_names, routs_count) = find_neighbor_indices(current_pos, grid_cells)
        print("=================================================")
        print("current_pos: ", current_pos, "  -  neighbor_indices: ", neighbor_indices, "  -  routs_count: ", routs_count, ")
        # print("current_pos: ", current_pos, "  -  neighbor_indices: ", neighbor_indices, "  -  routs_count: ", routs_count, "  -  last_pos: ", last_pos)
        # print("grid_bookings \n", grid_bookings)

        if (len(neighbor_indices) > 0):
            # pos = neighbor_indices[0]
            
            was_neighbor_with_earlier_node = False
            enforced_search = False
            open_routs_count += routs_count

            
            if (routs_count == 0):
                if (open_routs_count < 1):
                    return "Fail"
                else:
                    if (routs_count == 0):
                        enforced_search = True
                        print("enforced search")
            
            # last_pos = current_pos
            

            """
            if (last_pos == current_pos):

                # same_try_count += 1

                if (open_routs_count < 1):
                    return "Fail"
                else:
                    if (routs_count == 0):
                        enforced_search = True
                        print("enforced search")
            else:
                last_pos = current_pos
                # same_try_count = 0
            """
            
            
            
            for neighbor_pos, delta_name in zip(neighbor_indices, delta_names):
                
                # excluded_routes_count = 0
                
                # check if at least a route can be extracted
                if ( (grid_bookings[neighbor_pos[0], neighbor_pos[1]] == 0) or enforced_search):
                    
                    # check if at least a neighbor is already passed by this node
                    (check_neighbor_indices, check_delta_names, current_routs_count) = find_neighbor_indices(neighbor_pos, grid_cells)
                    
                    print("checking... neighbor_pos: ", neighbor_pos, "  -  check_neighbor_indices: ", check_neighbor_indices)
                    
                    for check_neighbor_pos, check_delta_name in zip(check_neighbor_indices, check_delta_names):
                        
                        if (current_pos == check_neighbor_pos):
                            continue
                        
                        if (grid_bookings[check_neighbor_pos[0], check_neighbor_pos[1]] == 1):
                            was_neighbor_with_earlier_node = True
                            route.remove_node_up_to(check_neighbor_pos)
                            
                            # open_routs_count -= 1
                            
                            # break

                    # route.append(neighbor_pos, delta_name, cost)
                    # grid_bookings[neighbor_pos[0], neighbor_pos[1]] = 1
                    
                    # route_continues = True
                    # current_pos = neighbor_pos
                            
                    # was not neighbor with any already-passed node
                    # if (not was_neighbor_with_earlier_node):
                    
                    route.append(neighbor_pos, delta_name, cost)
                    grid_bookings[neighbor_pos[0], neighbor_pos[1]] = 1
                    current_pos = neighbor_pos

                    open_routs_count -= 1

                    break
                    
        if (current_pos == goal_pos):
            branch_search_completed = True
            
        if (debug_counter == max_depth):
            branch_search_completed = True
        
        debug_counter += 1
    
    print("open_routs_count:", open_routs_count)
    
    return route
```
